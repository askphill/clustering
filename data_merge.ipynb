## Merge Shopify Customer and Segment Identifies

```python
df_master = pd.merge(shopify_customers, identities, left_on='ID', right_on='USER_ID', how='inner')
```

## Channel Data

```python
## GROUP BY USER_IDs AND THEN ADD ALL CHANNEL OCCURENCES IN DIFFERENT COLUMNS
per_user_channel = channel_user.groupby(['USER_ID', 'CHANNEL']).size().unstack(fill_value=0)
```

```python
## RENAME REMAINING COLUMNS TO IDENTIFY THEM AS CHANNELS
per_user_channel.columns = ['Channel_' + str(col).replace(' ', '_') for col in per_user_channel.columns]
```

```python
## TRANSFORM THE USER_ID INDEX TO A COLUMN
per_user_channel['USER_ID'] = per_user_channel.index
per_user_channel.reset_index(drop=True, inplace=True)
```

```python
df_master = pd.merge(df_master, per_user_channel, on='USER_ID', how='left')
```

## Event Type

```python
event_type.info()
```

```python
## GROUP BY USER_IDs AND THEN ADD ALL EVENT OCCURENCES IN DIFFERENT COLUMNS
event_group = event_type.groupby(['USER_ID', 'EVENT_TEXT']).size().unstack(fill_value=0)
```

```python
## RENAME REMAINING COLUMNS TO IDENTIFY THEM AS EVENTS
event_group.columns = ['Event_' + str(col).replace(' ', '_') for col in event_group.columns]
```

```python
## TRANSFORM THE USER_ID INDEX TO A COLUMN
event_group['USER_ID'] = event_group.index
event_group.reset_index(drop=True, inplace=True)
```

```python
event_group.head()
```

```python
df_master = pd.merge(df_master, event_group, on='USER_ID', how='left')
```
